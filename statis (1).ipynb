```python
# manual cal

pop_dat=[150,160,170,180,190]

# sample data

sample_dat=[157,172,180]

# caluclate population mean

pop_mean=sum(pop_dat)/len(pop_dat)

print(f'population mean is: {pop_mean}') #pop_mean


# calculate sample mean

sample_mean=sum(sample_dat)/len(sample_dat)

print(f'sample mean is: {sample_mean}') #sample_mean


```

```python
# using  numpy

#define dataset

import numpy as np

data=[85,90,78,92,88]

arith_mean=np.mean(data)

print(f'Arithmetic mean is: {arith_mean}')
```

```python
# using statistics

import statistics as sta

data=[85,90,78,92,88]

arith_mean=sta.mean(data)

print(f'Arithmetic mean is: {arith_mean}')
```
